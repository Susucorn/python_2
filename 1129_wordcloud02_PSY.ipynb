```python
# 라이브러리 불러오기

# 워드클라우드 생성을 위한 라이브러리
from wordcloud import WordCloud
# 그래프 및 시각화를 위한 라이브러리
import matplotlib.pyplot as plt
# 리스트 요소의 개수를 셀 때 사용하는 라이브러리
from collections import Counter
# 한국어 텍스트를 형태소 분석하기 위한 라이브러리
from konlpy.tag import Okt
# 배열 및 행렬 연산을 위한 라이브러리
import numpy as np      # as 내가 지정한 별칭
```

```python
okt = Okt() # 형태소 분석기 초기화
```

```python
# 텍스트 분석하기
# 파일 오픈
with open('대한민국헌법.txt', 'r', encoding='UTF-8') as f :
    text = f.read()

# 형태소 분석을 통해 명사 추출
nouns = okt.nouns(text)

# 한 글자 이상의 명사만 추출
words = [n for n in nouns if len(n) > 1]    # 길이가 1보다 큰 것만 골라 words 변수에 저장한다

# 단어의 빈도 수를 계산하자
c = Counter(words)

# 워드클라우드 생성을 위한 설정하기
# 코랩에서 제공하는 한글 폰트 사용을 위해 설정한다
#plt.rc('font', family = 'NanumGothic')
#font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'

font_path = 'gulim.ttc'
wc = WordCloud(width=400, height=400, font_path=font_path, max_font_size=250)  # font_path=font_path : 이름이 길기 때문에 폰드를 변수에 저장한다

# 단어의 빈도수에 따라 워드 클라우드 생성
gen = wc.generate_from_frequencies(c)   # c에 들어있는 워드 클라우드 내용을 담아 gen에 저장

# 워드클라우드 시각화 및 이미지 파일로 저장
plt.figure()
plt.imshow(gen)     # 변수에 담겨진 이미지를 보여준다
wc.to_file('헌법_클라우드.png')
```
